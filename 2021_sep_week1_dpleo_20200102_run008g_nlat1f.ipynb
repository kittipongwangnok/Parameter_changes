```python
data_Mass_ratio = pd.read_excel("2021_sep_week1_dpleo_20200102_run008g_nlat1f_Mass_ratio.xlsx", 
            names=["Mass_ratio_nlat1f", "Mass_ratio"]) 
# Preview the first 5 lines of the loaded data 
data_Mass_ratio.head()
```

```python
Mass_ratio_nlat1f = data_Mass_ratio["Mass_ratio_nlat1f"]
Mass_ratio = data_Mass_ratio["Mass_ratio"]


fig=plt.figure(figsize=(10, 5), tight_layout=True)
plt.tick_params(direction='in', which='both', bottom='on',top='on', right = 'on')
plt.plot(Mass_ratio_nlat1f, Mass_ratio, 'o', color='red', label = "Mass ratio")
#plt.plot(WD_Teff_nlat1f, WD_Teff, 'o', color='pink', label = "WD Teff")

plt.xlabel('nlat1f')
plt.ylabel('Mass ratio')

plt.grid(linestyle='dotted')
plt.legend()
plt.savefig("2021_sep_week1_dpleo_20200102_run008g_nlat1f_Mass_ratio.jpg")
plt.show()
```
